```python
import random
#movie name
a=['Haven','Limelight','Fear','Mermaids','Wings','Aliens','Argo','Goodfellas','Wolf','Devil','Seven','Malficient','Planes']



fig = [     """
                   --------
                   |      |
                   |      O
                   |    __|__
                   |      |
                   |     / \\
                   -
                """,
       """
                   --------
                   |      |
                   |      O
                   |    __|__
                   |      |
                   |     / 
                   -
                   """
                ,
       """
                   --------
                   |      |
                   |      O
                   |    __|__
                   |      |
                   |      
                   -
                 """
                  ,
       """
                   --------
                   |      |
                   |      O
                   |    __|
                   |      |
                   |     
                   -
                   """
                ,
       """

                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
          """
                ,
  """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                   """
                ,
       """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """
    ]


def play(mword,stepsLeft,gletters,done):
 while done:
   print(fig[7-stepsLeft-1])
   word=input(f" Tries Left= {stepsLeft} , Please enter   next letter:")

   if len(word)>1:
     print("Please enter a single character")
     continue
   if word in gletters:
     print("You have already entered this letter before, try something else")
     continue
   if not word.isalpha():
     print("Please enter a valid alphabet")
     continue

   gletters.append(word.lower())
   print ("Letters you have already guessed =>", end="")
   print(gletters)
   for letter in mword:
    if  letter.lower() in gletters:
       print(letter, end=" ")
    else:
       print('_',end=" ")
   print("")

   if word.lower() not in mword.lower():
     print("Oops , wrong guess")
     stepsLeft -= 1
   if stepsLeft==0:
     done=False
     break

   done=True;
   for letter in mword:
     if letter.lower() not in gletters:
       done=False

   if done:
       break;

   done=True;

 if done:
   print("Congrats, you guessed the word right!!!")

 else:
   print(fig[6])
   print("Better luck next time") 
   mword.replace(" ","")
   print(f"The correct answer was {mword}")  


def main():

  mword= random.choice(a)
  print(f"Its a {len(mword)} letter word")
  stepsLeft=6
  gletters=[]
  done = True
  play(mword,stepsLeft,gletters,done)

      
if __name__ == "__main__":
    main()
```
